```python
import cv2
from PIL import Image
from IPython.display import display

input_path = './Animal_images/1.jpg'
output_path = './BG_Removed_images/1.png'

# Load the input image
input_image = cv2.imread(input_path)

# Convert BGR image to RGB
input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)

# Apply a simple threshold to create a mask for background removal
_, mask = cv2.threshold(cv2.cvtColor(input_image, cv2.COLOR_RGB2GRAY), 240, 255, cv2.THRESH_BINARY_INV)

# Apply the mask to the original image
output_image = cv2.bitwise_and(input_image, input_image, mask=mask)

# Convert the NumPy array to PIL Image
output_image = Image.fromarray(output_image)

# Save the background-removed image
output_image.save(output_path)

# Display the original and background-removed images
display(Image.open(input_path))
display(output_image)
```
